```python
%%sql
SELECT *
  FROM facts
 LIMIT 5
```

```python
%%sql
SELECT MIN(population), 
       MAX(population), 
       MIN(population_growth), 
       MAX(population_growth)
  FROM facts
```

```python
%%sql
SELECT *
  FROM facts
  WHERE population == (SELECT MAX(population)
                        FROM facts
                     );
```

```python
%%sql
SELECT *
  FROM facts
  WHERE population == (SELECT MIN(population)
                        FROM facts
                     );
```

```python
%%sql
SELECT MIN(population), 
       MAX(population), 
       MIN(population_growth), 
       MAX(population_growth)
  FROM facts
 WHERE name <> 'World';
```

```python
%%sql
SELECT AVG(population), 
       AVG (area)
  FROM facts
 WHERE name <> 'World';
```

```python
%%sql
SELECT *
  FROM facts
Where population > (SELECT AVG(population) 
                    FROM facts
                    WHERE name <> 'World'
                   ) 
      AND area < (SELECT AVG(area) 
                  FROM facts
                  WHERE name <> 'World'
                 );
```

```python
%%sql
SELECT *
  FROM facts
  WHERE population == (SELECT MAX(population)
                        FROM facts
                       WHERE name <> 'World'
                     );
```

```python
%%sql
SELECT *
  FROM facts
  WHERE population_growth == (SELECT MAX(population_growth)
                        FROM facts
                       WHERE name <> 'World'
                     );
```

```python
%%sql
SELECT *, 
       CAST(area_water AS FLOAT) / area AS water_percentage
  FROM facts
  ORDER BY water_percentage DESC
  LIMIT 10;
```

```python
%%sql
SELECT *
  FROM facts
  WHERE area_land < area_water
```

```python
%%sql
SELECT *, 
       ROUND(population * (population_growth / 100)) AS expected_increase
  FROM facts
  WHERE name <> 'World'
  ORDER BY expected_increase DESC
  LIMIT 10;
```

```python
%%sql
SELECT *
  FROM facts
 WHERE death_rate > birth_rate;
```

```python
%%sql
SELECT *, 
       ROUND(CAST(population AS FLOAT) / area) AS pop_to_area
  FROM facts
  ORDER BY pop_to_area DESC
  LIMIT 10;
```
