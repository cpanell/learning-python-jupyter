```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]

star_wars.head()
```

```python
yes_no = {
    "Yes": True,
    "No": False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)

print(star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts())
print(star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts())

```

```python
column_names = star_wars.columns[3:9]
title_map = {
    'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    np.NaN: False
}

for col in column_names:
    star_wars[col] = star_wars[col].map(title_map)
    
for col in column_names:
    print (star_wars[col].value_counts())
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

star_wars = star_wars.rename(columns={
        'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.head()
```

```python
ranking_avg = {}
ranking_columns = star_wars.columns[9:15]

for col in ranking_columns:
    mean = star_wars[col].mean()
    ranking_avg[col] = mean
    
print(ranking_avg)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.bar(ranking_avg.keys(),ranking_avg.values())

```

```python
seen_sum = {}
seen_columns = star_wars.columns[3:9]

for col in seen_columns:
    sum = star_wars[col].sum()
    seen_sum[col] = sum
    
print(seen_sum)
```

```python
plt.bar(seen_sum.keys(),seen_sum.values())
```

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

star_trek_fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"] == "Yes"]
not_star_trek_fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"] == "No"]

star_wars_fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == True]
not_star_wars_fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == False]
```

```python
plt.bar(males.columns[9:15], males[males.columns[9:15]].mean())
plt.show()

plt.bar(females.columns[9:15], females[females.columns[9:15]].mean())
plt.show()
```

```python
plt.bar(males.columns[3:9], males[males.columns[3:9]].sum())
plt.show()

plt.bar(females.columns[3:9], females[females.columns[3:9]].sum())
plt.show()
```

```python
plt.bar(star_trek_fans.columns[9:15], star_trek_fans[star_trek_fans.columns[9:15]].mean())
plt.show()

plt.bar(not_star_trek_fans.columns[9:15], not_star_trek_fans[not_star_trek_fans.columns[9:15]].mean())
plt.show()
```

```python
plt.bar(star_trek_fans.columns[3:9], star_trek_fans[star_trek_fans.columns[3:9]].sum())
plt.show()

plt.bar(not_star_trek_fans.columns[3:9], not_star_trek_fans[not_star_trek_fans.columns[3:9]].sum())
plt.show()
```

```python
plt.bar(star_wars_fans.columns[9:15], star_wars_fans[star_wars_fans.columns[9:15]].mean())
plt.show()

plt.bar(not_star_wars_fans.columns[9:15], not_star_wars_fans[not_star_wars_fans.columns[9:15]].mean())
plt.show()
```
