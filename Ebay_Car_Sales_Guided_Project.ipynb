The goals of this project are to clean data scraped from ebay and analyze the included used car listings (and explore some unique benefits jupyter notebook provides for pandas).

```python
#Importing NumPy (as np) and pandas (as pd)
import numpy as np
import pandas as pd

autos = pd.read_csv('autos.csv', encoding = 'Latin-1')
```

We can observe from that the file includes 50,000 rows (a combination of strings and integers). We should convert some of the string columns (4 price and 11 odometer) to integers. We should convert the date/time columns using NumPy (7 yearOfRegistration, 12 monthOfRegistration, 16 dateCreated, 19 lastSeen). There are some NaN values we'll have to clean.

```python
#Cleaning up the column names.
autos_columns = autos.columns

autos_columns = autos_columns.str.replace('yearOfRegistration', 'registration_year')
autos_columns = autos_columns.str.replace('monthOfRegistration', 'registration_month')
autos_columns = autos_columns.str.replace('notRepairedDamage', 'unrepaired_damage')
autos_columns = autos_columns.str.replace('dateCreated', 'ad_created')
autos_columns = autos_columns.str.replace('dateCrawled ', 'date_crawled ')
autos_columns = autos_columns.str.replace('offerType', 'offer_type')
autos_columns = autos_columns.str.replace('vehicleType', 'vehicle_type')
autos_columns = autos_columns.str.replace('fuelType ', 'fuel_type ')
autos_columns = autos_columns.str.replace('nrOfPictures', 'number_of_pictures')
autos_columns = autos_columns.str.replace('postalCode', 'postal_code')
autos_columns = autos_columns.str.replace('lastSeen', 'last_seen')
autos_columns = autos_columns.str.replace('dateCrawled', 'date_crawled')
autos_columns = autos_columns.str.replace('powerPS', 'power_PS')
autos_columns = autos_columns.str.replace('fuelType', 'fuel_type')

autos.columns = autos_columns
print(autos.columns)
```

Above, I cleaned up the column names to conform with snakecase and clarify where necessary.

```python
autos.describe()
```

Observations:
* Column number_of_pictures can be removed (all 0s)
* Columns price and odometer are missing because they've been stored as text and need to be transformed.

```python
#Clean and convert the "price" and "odometer" columns to integers

autos['price'] = (autos['price']
                  .str.replace('$','')
                  .str.replace(',','')
                  .astype(int)
                 )
autos['odometer'] = (autos['odometer']
                     .str.replace(',','')
                     .str.replace('km','')
                     .astype(int)
                    )

autos.rename({'odometer':'odometer_km'}, axis=1, inplace=True)

autos.describe()
```

```python
#Analyzing the "price" and "odometer_km" columns

print("Price Description")
print(autos['price'].unique().shape)
print(autos['price'].describe())
print("Price Counts")
print(autos['price'].value_counts())

print("Odometer Description")
print(autos['odometer_km'].unique().shape)
print(autos['odometer_km'].describe())
print("Odometer Counts")
print(autos['odometer_km'].value_counts())
```

The range of values falls outside of what we would expect, some years are before cars are invented and some are in the future. We'll remove those. 

```python
#Only including cars registered between 1900 and 2016
autos = autos[autos['registration_year'].between(1900,2017)]
autos["registration_year"].value_counts(normalize=True).head(10)
```

```python
autos["brand"].value_counts(normalize=True).head(10)
```

```python
brand_counts = autos["brand"].value_counts(normalize=True).head(10)
top_brands = brand_counts[brand_counts > .05].index
print(top_brands)
```

```python
brand_mean_price = {}

for brand in top_brands:
    filter_brands = autos[autos['brand'] == brand]
    mean_price = filter_brands['price'].mean()
    brand_mean_price[brand] = int(mean_price)
    
print(brand_mean_price)
```

```python
#Convert dictionary to a panda dataframe
bmp_series = pd.Series(brand_mean_price)
brand_data = pd.DataFrame(bmp_series, columns=['mean_price'])
brand_data
```

```python
brand_mean_mileage = {}

for brand in top_brands:
    filter_brands = autos[autos['brand'] == brand]
    mean_mileage = filter_brands['price'].mean()
    brand_mean_mileage[brand] = int(mean_mileage)
    
brand_mean_mileage
```

```python
mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending=False)
mean_price = pd.Series(brand_mean_price).sort_values(ascending=False)

brand_data = pd.DataFrame(mean_mileage, columns=['mean_mileage'])
brand_data['mean price'] = mean_price

brand_data
```
