```python
#Importing Pandas and matplotlib and describing the data.

import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')

print(recent_grads.iloc[0])
print('\n')
print('Top Recent Grads:')
print(recent_grads.head(10))
print('\n')
print('Bottom Recent Grads:')
print(recent_grads.tail(10))
print('\n')
print('Describe Recent Grads:')
recent_grads.describe()
```

```python
#Removing na (and showing the change in row numbers)

raw_data_count = len(recent_grads)
print(raw_data_count)

recent_grads = recent_grads.dropna()

cleaned_data_count = len(recent_grads)
print(cleaned_data_count)
```

```python
ax1 = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax1.set_title('Median vs. Sample Size')
```

```python
ax2 = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax2.set_title('Unemployment Rate vs. Sample Size')
```

```python
ax3 = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax3.set_title('Median vs. Full Time')
```

```python
ax4 = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax4.set_title('Unemployment Rate vs. Share of Women')
```

```python
ax5 = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax5.set_title('Median vs. Men')
```

```python
ax6 = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax6.set_title('Median vs. Women')
```

```python
col = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']

fig = plt.figure(figsize=(8,36))

for r in range(0,7):
    ax = fig.add_subplot(8,1,r+1)
    ax = recent_grads[col[r]].plot(kind='hist', rot=30)
```
